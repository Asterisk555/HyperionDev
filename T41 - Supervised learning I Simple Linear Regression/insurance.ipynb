# Insurance analysis

```python
# Import required packages
import pandas as pd  # For database handling
import matplotlib.pyplot as plt  # For plotting
from sklearn import linear_model  # For linear regression model
```

```python
# Import insurance.csv
df = pd.read_csv('insurance.csv')
```

## Affect of Age on Insurance Costs

```python
# Plot a scatter plot with age on the x-axis and charges on the y-axis
plt.scatter(df['age'], df['charges'])
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Age vs Charges')
plt.show()
```

```python
# Using linear_model.LinearRegression() from sklearn, fit a model to the data

# Separate the (features/predictors/y) and the (target variable/response/x) in the dataset
X = df[['age']]
y = df['charges']

# Fit model to the data
lm = linear_model.LinearRegression()
model = lm.fit(X,y)

# Print the intercept and coefficient values of the linear regression model
print('Intercept:', model.intercept_)
print('Coefficient:', model.coef_)

# Make a prediction for a new customer who is 26 years old
new_age = [[26]]
prediction = model.predict(new_age)
print('Predicted charges:', prediction)
```

```python
# Plot another scatter plot with the best-fit line

# Plot the scatter plot
plt.scatter(X, y, color='blue')

# Plot the best-fit line
plt.plot(X, model.predict(X), color='red', linewidth=2)

# Set the labels and title
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Age vs Charges')
```
